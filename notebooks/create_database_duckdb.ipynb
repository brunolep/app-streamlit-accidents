# Création de la database duckdb

# Librairies

```python
import duckdb
import pandas as pd

```

# Chargement des CSV

```python
dv = pd.read_csv('../data/vehicules-2024.csv', sep = ";")
dl = pd.read_csv('../data/lieux-2024.csv', sep = ";")
da = pd.read_csv('../data/2024.csv', sep = ";")
du = pd.read_csv('../data/usagers-2024.csv', sep = ";")
```

# Nettoyage des données

# Creation du fichier DUCKDB

```python
dataframes_to_load = {
    "vehicules": dv,
    "localisations": dl,
    "accidents": da, # Nom de table suggéré pour 'da'
    "usagers": du
}
```

```python

# Nom du fichier DuckDB
duckdb_file = "../data/database_accidents_2024.duckdb" # J'ai renommé le fichier pour refléter l'année
```

```python

try:
    # Connexion à la base de données DuckDB (créera le fichier s'il n'existe pas)
    con = duckdb.connect(database=duckdb_file, read_only=False)

    for table_name, df in dataframes_to_load.items():
        print(f"Chargement du DataFrame pour la table '{table_name}'...")
        
        # Créer la table dans DuckDB et insérer les données du DataFrame
        # DuckDB peut directement ingérer un DataFrame pandas
        con.execute(f"CREATE OR REPLACE TABLE {table_name} AS SELECT * FROM df")
        print(f"Table '{table_name}' créée et remplie avec succès.")

    print(f"\nToutes les données ont été chargées dans '{duckdb_file}'.")

    # Optionnel : Vérifier les tables créées
    print("\nTables dans la base de données :")
    print(con.execute("SHOW TABLES;").fetchdf())

    # Optionnel : Afficher les 5 premières lignes d'une table (par exemple, 'accidents')
    # print("\nPremières lignes de la table 'accidents' :")
    # print(con.execute("SELECT * FROM accidents LIMIT 5;").fetchdf())

except Exception as e:
    print(f"Une erreur est survenue : {e}")
finally:
    # Fermer la connexion à la base de données
    if 'con' in locals() and con:
        con.close()
        print("Connexion DuckDB fermée.")
```
